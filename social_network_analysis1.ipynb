```
q=nx.Graph()
cities=['banglore','chennai','delhi','hyderabad','mumbbai','kolkata','pune']
for i in cities:
  q.add_node(i)
```

```
nx.draw(q,with_labels=True)
```

```
costs=[]
i=200
while i<2000:
  costs.append(i)
  i=i+200
```

```
costs
```

```
######### we are going to create 10 edges

while(q.number_of_edges()<10):                                                             # random.choice select one element from the list
  c1=random.choice(list(q.nodes()))
  c2=random.choice(list(q.nodes()))
  while c1 != c2 and q.has_edge(c1,c2)==0 :
    w=random.choice(costs)
    q.add_edge(c1,c2,weight=w)
```

```
nx.draw(q,with_labels=True)
```

```
k=nx.circular_layout(q)
nx.draw(q,k,with_labels=True)                        # if nodes are overlapping we can change layout nx.spectal_layout,circular_layout,spring_layout
```

```
k=nx.circular_layout(q)
nx.draw(q,k,with_labels=True)  
nx.draw_networkx_edge_labels(q,k)
```

```
nx.is_connected(q)                    # tells wheather it is  fullly connected
```

```
for i in q.nodes():
  for j in q.nodes():
    print(i,j,nx.has_path(q,i,j))
```

```
u='mumbbai'
v='delhi'
nx.dijkstra_path(q,u,v)                          # shortest path between mumbbai and delhi
```

```
u='mumbbai'
v='delhi'
nx.dijkstra_path_length(q,u,v)            # shortest pathlength between u and v
```

```
nx.single_source_dijkstra_path(q,'banglore')           # shortest path to every other node
                                                         # if there is no path the code will give exception
```
